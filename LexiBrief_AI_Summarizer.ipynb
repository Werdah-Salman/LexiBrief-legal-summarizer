```python
# ⚖️ Legal Document Summarizer using Hugging Face Transformers

# ✅ Step 1: Install the required library
!pip install transformers --quiet

# ✅ Step 2: Import the pipeline from Hugging Face
from transformers import pipeline

# ✅ Step 3: Load the summarization model
summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")

# ✅ Step 4: Define your legal text (example text)
legal_text = """
The plaintiff filed a lawsuit seeking damages for financial loss due to breach of contract.
After reviewing the evidence and hearing testimonies from both parties, the court found the defendant
guilty of failing to meet contractual obligations. The court awarded the plaintiff damages and legal costs.
"""

# ✅ Step 5: (Optional) Limit input length to avoid errors
max_length = 1024  # BART can handle ~1024 tokens
legal_text = legal_text[:max_length]

# ✅ Step 6: Summarize the text
summary = summarizer(legal_text, max_length=130, min_length=30, do_sample=False)

# ✅ Step 7: Display the summary
print("📄 Summary:\n")
print(summary[0]['summary_text'])
```
